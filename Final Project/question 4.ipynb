```python
def find2s(number):
    col = []
    for i in range(number, 0, -1):
        if (number - i >= 0 and i & (i - 1) == 0):
            number -= i
            col.append(i)
    return col
def FindMin(graph, graph2):
    number = len(graph2)
    finalgraph = []
    for i in range(number):
        col = []
        for j in range(number):
            list1 = graph[i]
            list2 = []
            for m in range(number):
                list2.append(graph2[m][j])
            numbermin = 0
            first = True
            for k in range(number):
                if list1[k] == 'inf' or list2[k] == 'inf':
                    continue
                elif list1[k] == 0 or list2[k] == 0:
                    continue
                else:
                    if (first):
                        numbermin = list1[k] + list2[k]
                        first = False
                    else:
                        numbermin = min(numbermin, list1[k]+list2[k])
            if numbermin == 0:
                col.append(float('inf'))
            else:
                col.append(numbermin)
        finalgraph.append(col)
    return finalgraph


def makemartix(graph, goal):
    if (goal % 2 == 0 and goal/2 >= 1):
        mm = makemartix(graph, int(goal/2))
        return FindMin(mm, mm)
    else:
        return graph

print("how many vertices do you have ?")
number = int(input())
print()
print("write inf for no path between 2 vertices")
graph = []
for i in range(number):
    vorodi = input().split(" ")
    graph.append(vorodi)
for i in range(number):
    for j in range(number):
        if(graph[i][j] == "inf"):
            graph[i][j] = float('inf')
        else:
            graph[i][j]  = int(graph[i][j])
print("graph is :")
print(graph)
print()
print()
print("graph martabe chand ?")
tool = int(input())
array = find2s(tool)
array.reverse()
storage = {}
for i in range(len(array)):
    storage[i] = makemartix(graph, array[i])
hasel = storage[0]
for i in range(1,len(storage)):
    hasel = FindMin(storage[i], hasel)
print(hasel)
```
